```python
Data1 = BSData('PWM500_BAR_CENTERED.txt')
Data1.read_data()
Data2 = BSData('PWM500_BAR_2CMR.txt')
Data2.read_data()
```

```python
direction_data1 = pd.Series(data=list(Data1.data["Dir"]), dtype=bool)
data_data1 = pd.Series(data=list(Data1.data["Current"]), dtype=float)
data1_dir1 = data_data1[direction_data1]
data1_dir0 = data_data1[direction_data1 == False]
```

```python
direction_data2 = pd.Series(data=list(Data2.data["Dir"]), dtype=bool)
data_data2 = pd.Series(data=list(Data2.data["Current"]), dtype=float)
data2_dir1 = data_data2[direction_data2]
data2_dir0 = data_data2[direction_data2 == False]
```

```python
print(f'data1: dir1 mean: {data2_dir1.describe()["mean"]} dir0 mean: {data2_dir0.describe()["mean"]}')
print(f'data2: dir1 mean: {data1_dir1.describe()["mean"]} dir0 mean: {data1_dir0.describe()["mean"]}')
print(f'Diff: dir1 mean: {data2_dir1.describe()["mean"] - data1_dir1.describe()["mean"]} '
      f'dir0 mean: {data2_dir0.describe()["mean"] - data1_dir0.describe()["mean"]}')
```

```python
print(f'data1: dir1 std: {data2_dir1.describe()["std"]} dir0 std: {data2_dir0.describe()["std"]}')
print(f'data2: dir1 std: {data1_dir1.describe()["std"]} dir0 std: {data1_dir0.describe()["std"]}')
print(f'Diff: dir1 std: {data2_dir1.describe()["std"] - data1_dir1.describe()["std"]} '
      f'dir0 std: {data2_dir0.describe()["std"] - data1_dir0.describe()["std"]}')
```
